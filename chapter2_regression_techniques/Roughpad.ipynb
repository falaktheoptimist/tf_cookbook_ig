```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(-5,5,0.2)
```

```python
y1 = 2*x
y2 = 2*x+3
y3 = 2*x-5
```

```python
%matplotlib inline
#plt.figure(figsize=[8,5]);
fig, ax = plt.subplots(figsize=[8,5])

ax.grid()

ax.plot(x,y1,x,y2, x, y3)
#ax.set_aspect('equal')
ax.legend(['y=2x','y=2x+3','y=2x-5'])
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
```
